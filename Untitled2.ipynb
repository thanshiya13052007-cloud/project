```python
import pandas as pd
import matplotlib.pyplot as plt
data={"Order_Time": ["Morning", "Afternoon", "Evening", "Morning", "Night", "Evening", "Afternoon" ],
      "Food_Item": ["Pizza", "Burger", "Dosa", "Pizza", "Biriyani", "Burger", "Pizza"],
      "Location": ["Chennai", "Madurai", "Coimbatore", "Chennai", "Madurai", "Chennai", "Coimmbatore"]
      }
df = pd.DataFrame(data)
df.to_csv("food_delivery.csv",
          index=False)
print("CSV file created successfully\n")
df = pd.read_csv("food_delivery.csv")
df = df.dropna().drop_duplicates()
chennai_orders = df[df["Location"] == "Chennai"]
print("Filtered Data (Chennai Orders): ")
print(chennai_orders)
print("Peak Orders Times: ")
print(df["Order_Time"].value_counts())
print("\nMost Popular Dishes: ")
print(df["Food_Item"].value_counts())
print("\nOrders by Location: ")
print(df["Location"].value_counts())
print("\nFood Delivery Sales & customer Insights Analysis Completed")
```
