```python
import pandas as pd
import matplotlib.pyplot as plt
data={"Order_Time": ["Morning", "Afternoon", "Eveninng", "Night", "Morning", "Evening", "Afternoon", "Night", "Evening", "Morning"],
      "Location": ["Chennai", "Madurai", "Coimbatore", "Chennai", "Madurai", "Chennai", "Coimbatore", "Madurai", "Chennai", "Coimbatore"],
      "Food_Item": ["Pizza", "Burger", "Dosa", "Biryani", "Pizza", "Burger", "Pizza", "Biryani", "Pizza","Dosa"]
      }
df = pd.DataFrame(data)
print("Food Delivery Dataset")
print(df)
print("\n--- Order Analysis ---")
print("\nOrders by Time: ")
print(df["Order_Time"].value_counts())
print("\nOrders by Location")
print(df["Location"].value_counts())
print("\nOrders by Food Item: ")
print(df["Food_Item"].value_counts())
peak_time=df["Order_Time"].value_counts().idxmax()
print("\n--- Peak Orders Time ---")
print("Peak Order Time: ", peak_time)
popular_food=df["Food_Item"].value_counts().idxmax()
print("\n--- Most Popular Dish ---")
print("Most Popular Dish: ", popular_food)
print("\nOrders from Chennai ---")
print(df[df["Location"] == "Chennai"])
df["Food_Item"].value_counts().plot(kind="bar")
plt.title("Popular Food Items")
plt.xlabel("Food Item")
plt.ylabel("Number of Orders")
plt.show()
df["Location"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Orders by Location")
plt.ylabel("")
plt.show()
```
